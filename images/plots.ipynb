```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def equals(row):
    if row['Predicted'] == row['Actual']:
        return 1
    else:
        return 0
```

## Shuffle / Non-Shuffle Analysis on 33% Split

```python
df1_ns_vs33 = pd.read_csv('../predictions/20191101-1232-X1_NS_VS33.csv')
df2_ns_vs33 = pd.read_csv('../predictions/20191101-1235-X2_NS_VS33.csv')
df3_ns_vs33 = pd.read_csv('../predictions/20191101-1243-X3_NS_VS33.csv')

df1_s_vs33 = pd.read_csv('../predictions/20191101-1255-X1_S_VS33.csv')
df2_s_vs33 = pd.read_csv('../predictions/20191101-1258-X2_S_VS33.csv')
df3_s_vs33 = pd.read_csv('../predictions/20191101-1302-X3_S_VS33.csv')
```

```python
df1_ns_vs33['Correct_Pred'] = df1_ns_vs33.apply(lambda row: equals(row), axis=1)
df2_ns_vs33['Correct_Pred'] = df2_ns_vs33.apply(lambda row: equals(row), axis=1)
df3_ns_vs33['Correct_Pred'] = df3_ns_vs33.apply(lambda row: equals(row), axis=1)

df1_s_vs33['Correct_Pred'] = df1_s_vs33.apply(lambda row: equals(row), axis=1)
df2_s_vs33['Correct_Pred'] = df2_s_vs33.apply(lambda row: equals(row), axis=1)
df3_s_vs33['Correct_Pred'] = df3_s_vs33.apply(lambda row: equals(row), axis=1)
```

```python
df1_ns_vs33_nr = df1_ns_vs33[df1_ns_vs33['Actual'] == 0]
df1_ns_vs33_r = df1_ns_vs33[df1_ns_vs33['Actual'] == 1]

df2_ns_vs33_nr = df2_ns_vs33[df2_ns_vs33['Actual'] == 0]
df2_ns_vs33_r = df2_ns_vs33[df2_ns_vs33['Actual'] == 1]

df3_ns_vs33_nr = df3_ns_vs33[df3_ns_vs33['Actual'] == 0]
df3_ns_vs33_r = df3_ns_vs33[df3_ns_vs33['Actual'] == 1]

df1_s_vs33_nr = df1_s_vs33[df1_s_vs33['Actual'] == 0]
df1_s_vs33_r = df1_s_vs33[df1_s_vs33['Actual'] == 1]

df2_s_vs33_nr = df2_s_vs33[df2_s_vs33['Actual'] == 0]
df2_s_vs33_r = df2_s_vs33[df2_s_vs33['Actual'] == 1]

df3_s_vs33_nr = df3_s_vs33[df3_s_vs33['Actual'] == 0]
df3_s_vs33_r = df3_s_vs33[df3_s_vs33['Actual'] == 1]
```

```python
recession_ns_33 = [
    df1_ns_vs33_r['Correct_Pred'].mean(),
    df2_ns_vs33_r['Correct_Pred'].mean(),
    df3_ns_vs33_r['Correct_Pred'].mean()
]

no_recession_ns_33 = [
    df1_ns_vs33_nr['Correct_Pred'].mean(),
    df2_ns_vs33_nr['Correct_Pred'].mean(),
    df3_ns_vs33_nr['Correct_Pred'].mean()
]

recession_s_33 = [
    df1_s_vs33_r['Correct_Pred'].mean(),
    df2_s_vs33_r['Correct_Pred'].mean(),
    df3_s_vs33_r['Correct_Pred'].mean()
]

no_recession_s_33 = [
    df1_s_vs33_nr['Correct_Pred'].mean(),
    df2_s_vs33_nr['Correct_Pred'].mean(),
    df3_s_vs33_nr['Correct_Pred'].mean()
]
```

```python
fig, ax = plt.subplots(figsize=(15,6))

x_labels = ['1-Qtr Out', '2-Qtrs Out', '4-Qtrs Out']

graph_parameters = {
    'No-Shuffle: recession':{'y_data':recession_ns_33,'color':'k','marker':'o', 'linestyle':'--'},
    'No-Shuffle: no recession':{'y_data':no_recession_ns_33,'color':'k','marker':'o', 'linestyle':'-'},
    'Shuffle: recession':{'y_data':recession_s_33,'color':'r','marker':'o', 'linestyle':'--'},
    'Shuffle: no recession':{'y_data':no_recession_s_33,'color':'r','marker':'o', 'linestyle':'-'},
}

for subset,params in graph_parameters.items():
    y = params['y_data']
    
    ax.plot(x_labels, y, color=params['color'], marker=params['marker'], linestyle=params['linestyle'])
    
    for x,y in zip(x_labels,y):

        label = "{:.2f}".format(y)

        plt.annotate(label, # this is the text
                     (x,y), # this is the point to label
                     textcoords="offset points", # how to position the text
                     xytext=(0,10), # distance from text to points (x,y)
                     ha='center') # horizontal alignment can be left, right or center

plt.ylim(top=1)  # adjust the top leaving bottom unchanged

# Set Labels
ax.set_xlabel('Models', fontsize='large')
ax.set_ylabel('Pct Accuracy on Actual Recession/No Recession', fontsize='large')
ax.set_title('Shuffle vs. No-Shuffle\n33% Validation Split', fontsize='x-large', fontweight='bold')

plt.legend(graph_parameters.keys(), loc='best')

ax.grid()
plt.savefig("./plots/shuffle_noShuffule_vs33.png")

plt.show()
```
